```python
# by default it considers few values as NA, but I want to it consider Not-Available, unkowon as NA only

df1 = pd.read_csv('orders_data.csv', na_values=['Not Available', 'unknown'])
df1['Ship Mode'].unique()
```

```python
df1.head()
```

#### Add new necessary columns discount, sale price and profit

```python
df1['Selling Price'] = df1['List Price'] - df1['List Price'] * (df1['Discount Percent'] / 100)
df1['Selling Price']
```

```python
df1.head(2)
```

```python
df1['Profit'] = (df1['Selling Price'] - df1['cost price'])
df1['Profit']
```

```python
df1.head(2)
```

```python
df1.rename(columns={'Profit': 'Unit Profit', 'Selling Price': 'Unit Selling Price'}, inplace= True)
df1.head()
```

```python
df1.dtypes
```

```python
df1.head(2)
```

#### Change datatype of Order Date column from object to datetime

```python
df1['Order Date'] = pd.to_datetime(df1['Order Date'], format='%d-%m-%Y')
```

```python
df1.dtypes
```

```python
df1.head()
```

#### Drop few unnecessary columns

```python
df1.drop(columns=['cost price', 'List Price', 'Discount Percent'], inplace=True)
```

```python
df1.head(2)
```

```python
df1['Total Profit'] = df1['Quantity'] * df1['Unit Profit']
```

```python
df1.head(2)
```

### Q13. Find month over month growth comparison for 2022 and 2023 sales eg : jan 2022 vs jan 2023.

```python
df1.head(2)
```

```python
df1['Year'] = df1['Order Date'].dt.year
```

```python
df1['Quarter'] = df1['Order Date'].dt.quarter
```

```python
df1.head(2)
```

```python
df1.groupby(['Month', 'Year'])
```

```python
a = df1.groupby(['Month', 'Year']).sum('Total Sales')['Total Sales'].reset_index()
a
```

```python
pivotTable = a.pivot_table(index='Month', columns='Year', values='Total Sales', aggfunc='sum')
pivotTable
```

```python
comparedTable = pivotTable.reset_index()
comparedTable
```

```python
comparedTable['YtoY Growth'] = comparedTable[2023] - comparedTable[2022]
comparedTable
```

```python
comparedTable['YtoY Growth percentage'] = comparedTable['YtoY Growth'] / comparedTable[2022]
comparedTable['YtoY Growth percentage'] = comparedTable['YtoY Growth percentage'].apply(lambda x: "{:.2%}".format(x))
comparedTable
```

### Q15. Which sub category had highest growth by profit in 2023 compare to 2022.

```python
df1.head(2)
```

```python
m = df1.groupby(['Sub Category', 'Year']).sum(['Total Profit'])['Total Profit'].reset_index()
m
```

```python
n = m.pivot_table(index= 'Sub Category', columns= 'Year', values= 'Total Profit')
n
```

```python
n['diff'] = n[2023] - n[2022]
nSorted = n.sort_values(by=['diff'], ascending= False)
nSorted.head(1)
```
